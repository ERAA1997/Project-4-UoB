```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import pandas as pd

# Assuming 'data' is your DataFrame and 'Potability' is the target variable
X = cleaned_df.drop('Potability', axis=1)
y = cleaned_df['Potability']

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Applying SMOTE to the training data
smote = SMOTE(random_state=1)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Now, X_train_resampled and y_train_resampled are your new training set
```

```python
# Initialize the RandomForestClassifier
rf_classifier = RandomForestClassifier(random_state=1)
```

```python
# Define the grid of parameters to search over
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
# Setup GridSearchCV
grid_search = GridSearchCV(estimator=rf_classifier, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)

# Fit GridSearchCV to the resampled training data
grid_search.fit(X_train_resampled, y_train_resampled)
```

```python
# Best parameters and accuracy
print("Best parameters found: ", grid_search.best_params_)
print("Best cross-validated accuracy: ", grid_search.best_score_)
```

```python
# Evaluate on the test set (without resampling)
best_estimator = grid_search.best_estimator_
y_pred = best_estimator.predict(X_test)
test_accuracy = accuracy_score(y_test, y_pred)   
```

```python
print(f"Test accuracy: {test_accuracy}")
print(classification_report(y_test, y_pred))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Specify the columns to plot against 'Sulfate'
columns_to_plot = cleaned_df.columns[:-1]  # Exclude the 'Potability' column
```

```python
# Set up the figure and axes
num_plots = len(columns_to_plot)
fig, axes = plt.subplots(nrows=num_plots, figsize=(8, 6*num_plots))

# Loop through each column and create pair plots
for i, column in enumerate(columns_to_plot):
    sns.scatterplot(x=column, y='Sulfate', hue='Potability', data=cleaned_df, ax=axes[i])
    axes[i].set_title(f'Pair Plot of {column} vs Sulfate')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Sulfate')
    axes[i].legend()

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Linear Regression on the 'Sulfate' Column
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Extracting the 'Sulfate' column
X_sulfate = cleaned_df[['Sulfate']]
y_potability = cleaned_df['Potability']
```

```python
# Initialize and fit the linear regression model
regression_model = LinearRegression()
regression_model.fit(X_sulfate, y_potability)
```

```python
# Visualize the linear regression line
plt.figure(figsize=(10, 6))
plt.scatter(y_potability, X_sulfate, color='blue', label='Actual Data')  # Swap x and y
plt.plot(regression_model.predict(X_sulfate), X_sulfate, color='red', label='Linear Regression')  # Swap x and y
plt.title('Linear Regression on Sulfate Column')
plt.xlabel('Potability')  # Swap x and y
plt.ylabel('Sulfate')  # Swap x and y
plt.legend()
plt.show()
```
